```python
#sort by values
df.sort_values(by='Region')
```

```python
# Group data by region and calculate total sales
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

# Sort regions by sales
region_sales = region_sales.sort_values(by='Sales', ascending=False)
```

# **Visualization**

```python
# Plotting the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=region_sales, palette='viridis')

# Adding labels and title
plt.title('Total Sales by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Pie chart for sales by region
plt.figure(figsize=(8, 8))
plt.pie(region_sales['Sales'], labels=region_sales['Region'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(region_sales)))
plt.title('Sales Distribution by Region', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

```python
# Horizontal Bar Plot for Sales by Region
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales', y='Region', data=region_sales, palette='viridis')

# Adding labels and title
plt.title('Total Sales by Region', fontsize=16)
plt.xlabel('Total Sales', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Boxplot to show sales distribution by Region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Sales', data=df, palette='viridis')

# Adding labels and title
plt.title('Sales Distribution by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Violin plot for sales by Region and City
plt.figure(figsize=(14, 8))
sns.violinplot(x='Region', y='Sales', data=df, palette='viridis')

# Adding labels and title
plt.title('Sales Distribution by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Convert 'Order Date' to datetime if it's not already
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y') # Changed format to day-month-year

# Grouping data by 'Order Date' to get total sales over time
sales_by_date = df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python

# Plotting the line graph for sales over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Date', y='Sales', data=sales_by_date, marker='o', color='b')

# Adding labels and title
plt.title('Sales Over Time (Line Graph)', fontsize=16)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plotting the histogram for sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30, kde=True, color='g')

# Adding labels and title
plt.title('Sales Distribution (Histogram)', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```

```python
### Heatmap
# Create a correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlations', fontsize=16)
plt.show()
```

```python
### Regression Plot
# Regression plot for Sales vs Postal Code
plt.figure(figsize=(8, 6))
sns.regplot(x='Postal Code', y='Sales', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Regression Plot: Postal Code vs Sales', fontsize=16)
plt.xlabel('Postal Code', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.show()
```

```python
# Grouping data by 'Region' and 'Order Date' to calculate total sales
sales_by_region = df.groupby(['Region', 'Order Date'])['Sales'].sum().reset_index()

# Line plot for total sales by region over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Date', y='Sales', hue='Region', data=sales_by_region, marker='o')

# Adding labels and title
plt.title('Sales Over Time by Region', fontsize=16)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
#Histogram of total sales by region
plt.figure(figsize=(10, 6))

# Plot histogram
sns.histplot(data=df, x='Sales', hue='Region', kde=True, palette='husl', bins=20, multiple='stack')

# Adding labels and title
plt.title('Distribution of Sales by Region', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```

```python
#Create a rug plot for 'Sales' grouped by 'Region'
plt.figure(figsize=(10, 6))
sns.rugplot(data=df, x='Sales', hue='Region', height=0.05, palette='viridis')

# Adding labels and title
plt.title('Rug Plot of Sales by Region', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Density Indicator', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Pair Plot: Numerical columns and Region as hue
sns.pairplot(df, vars=['Sales', 'Postal Code'], hue='Region', diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Sales and Postal Code by Region', y=1.02, fontsize=16)
plt.show()
```

```python
# Group data by Region and Category
region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()

# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', hue='Category', data=region_category_sales, palette='viridis')

# Adding labels and title
plt.title('Total Sales by Region and Category', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# KDE Plot: Sales distribution by Region
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Sales', hue='Region', fill=True, palette='husl', alpha=0.6)

# Adding labels and title
plt.title('KDE Plot of Sales by Region', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
plt.show()
```

#Conclusion:
Analysis of geographic trends reveals that the South region dominates sales across all categories, with consistent performance. The West region shows high variability in sales, indicating opportunities for growth in underperforming sub-regions. Furniture sales in the Central region lag significantly, requiring targeted marketing efforts. Insights from KDE plots suggest that high-value sales are concentrated in the South, making it a key focus for premium product strategies. Overall, leveraging strengths in high-performing regions and addressing gaps in weaker areas can enhance sales performance across the board.
